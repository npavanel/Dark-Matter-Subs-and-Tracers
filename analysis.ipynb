An error in spicy ramen has created redundant copies of per scenario information. All information for each scenario is now outputted in a single data file of length 65. The first 5 elements are for the entire scenario. The next 20 elements are for the first tracer. The 20 after that for the next. Etc.

    - [0-4]: total subhalos[0], mass_dif[1], #10e6 subs[2], #10e7 subs[3], #10e8 subs[4]
    
    - In general each tracer has structure: [i-i+3] -> [final,min,max,std]
    - For tracer one: vx[5-8], vy[9-12], vz[13-16], r[17-20], E[21-24]
    - For other tracers add 20

### Imports, scaling relations, time domain

### The function that determines the dissolution time ratio is below:

```python
from scipy.stats import median_absolute_deviation as mad
#define a function to return the ratio of tdiss(Rrandom)/tdiss(Rgc)
def t_diss(smf,r_tracer, r_data, time_domain):
    
    tdiss = [] # initialize a list for output
    
    sigTdiss = []
    
    smf=smf # define the substructure mass fraction
    sp = gp.LogarithmicHaloPotential(amp=1.-smf, ro=ro,vo=vo) # define the smooth potential
    
    if type(r_tracer) == list:
        
        for r in range(len(r_tracer)): # loop through all tracers
            
            tdiss_r = []
            
            tdiss_rtracer = r_tracer[r] * (220/sp.vcirc(r_tracer[r])) # compute the dissolution time for the circ tracer
            
            for t in range(len(time_domain)): # loop through every timestep, computing the dissolution ratio at each timestep
            
                r_random = np.random.normal(r_tracer[r],r_data[r][t],50000) # compute a random distribution of 1000 radii with width of sigR
                tdiss_random = r_random * (220/sp.vcirc(r_random)) # compute the dissolution time for the random radii 
                tdiss_ratio = tdiss_random / tdiss_rtracer # compute the dissolution ratio for the timestep
                tdiss_r.append(tdiss_ratio) # append the dissolution ratio for the timestep to the overall list 
                #if t == 249:
                    #print('Tracer',r,'tdiss_ratio std:',np.std(tdiss_ratio))
        
            sigTdiss.append(mad(tdiss_r,axis=1)) # append the distribution of tdiss at each timestep for a tracer to an output list
        
        return sigTdiss[0], sigTdiss[1], sigTdiss[2], sigTdiss[3], sigTdiss[4] # return each tracer's sigmat diss
```
